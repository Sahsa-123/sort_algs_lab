```python
import pandas as pd
import matplotlib.pyplot as plt
import os

sort_algs = ["bubble", "shaker", "heap", "std"]

data = {}
for sort in sort_algs:
    path = os.path.join(f"outputs_{sort}", "benchmarks.csv")
    if os.path.exists(path):
        df = pd.read_csv(path)
        df = df.sort_values(by='len')
        data[sort] = df
    else:
        print(f"  : {path}")


plt.figure(figsize=(10, 6))
for sort, df in data.items():
    plt.plot(df['len'], df['time'], label=sort, marker='o')

plt.title("     ")
plt.xlabel(" ")
plt.ylabel(" ()")
plt.yscale('log') 
plt.legend()
plt.grid(True)
plt.show()
```
